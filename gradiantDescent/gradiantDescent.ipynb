# Notebook imports and packages

```python
from matplotlib import pyplot as plt
import  numpy as np
%matplotlib inline
```

# Example 1: A simple cost function
$f(x) = x^2 + x + 1$

```python
def cost_func_1(x: float) -> float:
    return (x**2) + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=100)
```

## Slopes and derivatives

```python
def df(x):
    return (2*x) + 1
```

## Python loops and gradian descent

```python
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.00001

x_values = [new_x]
slopes = [df(new_x)]

for n in range(500):
    prev_x = new_x
    gradient = df(prev_x)

    # Stepping down the slope
    new_x = prev_x - (step_multiplier * gradient)
    
    # Add our slope and x values into their lists
    x_values.append(new_x)
    slopes.append(df(new_x))

    step_size = abs(new_x - prev_x)
    if step_size < precision:
        print(f"Total number of steps: {n}")
        break

print(f"Local Minimum occurs at {new_x}")
print(f"Slope at this point is {df(new_x)}")
print(f"Cost at this point is {cost_func_1(new_x)}")
```

```python
# superimpose gradient descent on Cost function plot and slope of the cost function
plt.figure(figsize=[20,5])


# First chart: Cost function
plt.subplot(1, 3, 1)
plt.title("Slope of cost function")
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel("X", fontsize=16)
plt.ylabel("f(X)", fontsize=16)
plt.plot(x_1, cost_func_1(x_1), color="blue", linewidth=3)

np_x_values = np.array(x_values)
plt.scatter(x_values, cost_func_1(np_x_values), color="red", s=100, alpha=0.6)

# Second chart of the derivative
plt.subplot(1,3,2)
plt.title("Slope of cost function")
plt.grid()
plt.xlim([-2,3])
plt.ylim([-3,6])
plt.xlabel("X", fontsize=16)
plt.ylabel("df(X)", fontsize=16)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5)

plt.scatter(x_values, slopes, color="red", s=100, alpha=0.6)

# Third chart of the gradient descent
plt.subplot(1,3,3)
plt.title("Gradient Descent")
plt.grid()
plt.xlim([-.55,.2])
plt.ylim([-.2,.8])
plt.xlabel("X", fontsize=16)
plt.ylabel("df(X)", fontsize=16)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5)

plt.scatter(x_values, slopes, color="red", s=100, alpha=0.6)

plt.show()
```
