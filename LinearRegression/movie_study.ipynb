```python
import pandas as pd
from pandas import DataFrame
from matplotlib import pyplot as plt
data = pd.read_csv("cost_revenue.csv")
data.describe()
```

```python
x_axis = DataFrame(data, columns=["production_budget_usd"])
y_axis = DataFrame(data, columns=["world_wide_gross_usd"])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_axis, y_axis)
plt.title("Film Cost vs Global Rev",)
plt.xlabel("Production Cost ($)")
plt.ylabel("World wide gross($)")
plt.ylim(0, 3e9)
plt.xlim(0, 5e8)
plt.show()
```
